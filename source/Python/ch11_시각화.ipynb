`````python
# 시작전 설정
import matplotlib.pyplot as plt
%matplotlib inline
# 그래프 해상도 높임
%config InlineBackend.figure_format='retina'
#한글설정
plt.rc('font', family='Malgun Gothic') # window os
#plt.rc('font', family='AppleGothic') #mac os
plt.rc('axes', unicode_minus=False)
# 경고 메세지 안보이게
import warnings
warnings.filterwarnings(action='ignore')
`````

`````python
import matplotlib.pyplot as plt
`````

`````python
import seaborn as sns 
iris = sns.load_dataset("iris")
iris.shape
`````

`````python
iris.plot(kind="line", figsize=(12,4))
plt.legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)
plt.show()
`````

`````python
iris.plot(kind="box", figsize=(12,4))
`````

`````python
plt.figure(figsize = (10,2))
plt.plot(iris.petal_length, iris.petal_width, '.')  # '-'으로하면 선 그래프
plt.title("두변수간 scatter plot")
plt.xlabel("iris.petal_length 값")
plt.ylabel("iris.petal_width 값")
plt.show()
`````

# 3절 seaborn
- matplotlib을 기반으로 만든 고수준 그래픽 라이브러리
• 공식사이트 : https://seaborn.pydata.org
• seaborn API : https://seaborn.pydata.org/api.html
- Seaborn으로 그래프를 그리기 위해서 다음 단계를따릅
```다.  
1)데이터 준비
2) 미적속성 설정
3) 함수를 이용하여 그래프 그리기
4) 그래 ````출력, 저장


`````python
plt.figure(figsize = (10,4))
sns.scatterplot(data=iris, x="petal_length", y = "petal_width", hue = "species")  # hue는 범례별로 구분해줌
plt.show()
`````

`````python
plt.figure(figsize = (10,4))
sns.barplot(data=iris, x="species", y = "petal_width")  # hue는 범례별로 구분해줌
plt.show()
`````

`````python
plt.figure(figsize = (10,4))
sns.lmplot(data=iris, x="petal_length", y = "petal_width")  # 회귀선 그려줌
plt.show()
`````

`````python
plt.figure(figsize = (10,4))
sns.lmplot(data=iris, x="petal_length", y = "petal_width", hue = "species")  # 회귀선 종별로그려줌
plt.show()
`````

`````python
plt.figure(figsize = (10,4))
sns.boxplot(data=iris, y = "petal_length", hue = "species") 
plt.show()
`````

`````python
plt.figure(figsize = (10,4))
sns.boxenplot(data=iris, y = "petal_length", hue = "species") 
plt.show()
`````

`````python
plt.figure(figsize = (10,4))
sns.violinplot(data=iris, y = "petal_length", hue = "species")   # 분포까지 그려줌
plt.show()
`````
